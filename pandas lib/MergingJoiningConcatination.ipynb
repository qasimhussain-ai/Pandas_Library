```python
import numpy as np
import pandas as pd
```

```python
# Mergeing 2 Datafreme
```

```python
employees = pd.DataFrame({
    'Emp_id':[1,2,3,4,5],
    'Emp_name':['Qasim','Ali','Hassan','Mouzan','Awais'],
    'Emp_dep':['BSCS','PHY','IT','CHE','Finance']
    
})

salaries = pd.DataFrame({
    'Emp_id':[1,2,3,6,7],
    'salary':[90000,80000,70000,60000,50000],
    'bnous':[5000,6000,7000,8000,9000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='Emp_id',how='inner')
```

```python
pd.merge(employees,salaries,on='Emp_id',how='outer')
```

```python
pd.merge(employees,salaries,on='Emp_id',how='left')
```

```python
pd.merge(employees,salaries,on='Emp_id',how='right')
```

```python
# Concatination of 2 Datafreame
```

```python
df1 = pd.DataFrame({
    'A':['A1','A2','A3'], 
    'B':['B1','B2','B3'],
    'C':['C1','C2','C3'] 
})

df2 = pd.DataFrame({
    'A':['A4','A5','A6'], 
    'B':['B4','B5','B6'],
    'C':['C4','C5','C6'] 
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

Joining of 2 DATAFRAME

```python
df1 = pd.DataFrame({
    'name':['Qasim','Ali','Mouzan'], 
},index=[1,2,3])

df2 = pd.DataFrame({
    'score':[85,50,75]
},index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1,how='outer')
```
