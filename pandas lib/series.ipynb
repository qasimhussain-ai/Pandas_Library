```python
# Pandas series
# A series is a one_dimensional labeled array capable of handling any data type . The axis lables are collectively called index
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python

pd.Series(arr)
```

```python
pd.Series(d)
```
