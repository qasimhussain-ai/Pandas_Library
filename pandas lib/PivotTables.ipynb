```python
import numpy as np 
import pandas as pd 
```

```python
# Pivot Table
```

```python
data = {
    'date':pd.date_range('2025-10-29',periods=20), 
    'Product':['A','B','C','D'] * 5, 
    'Region':['East','Weat','North','South','East','Weat','North','South','East','Weat','North','South','East','Weat','North','South','East','Weat','North','South'], 
    'Sales':np.random.randint(100,1000,20), 
    'Units':np.random.randint(10,100,20), 
    'Rep':['Qasim','Ali','Mouzan','Subhan','Qasim','Ali','Mouzan','Subhan','Qasim','Ali','Mouzan','Subhan','Qasim','Ali','Mouzan','Subhan','Qasim','Ali','Mouzan','Subhan']
    
}

df = pd.DataFrame(data)

df['Month'] = df['date'].dt.month_name()
df['Quarter'] = 'Q' + df['date'].dt.quarter.astype(str)

df
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median')
```

```python
pivot2 = pd.pivot_table(df,values=['Sales','Units'],index='Region',columns = 'Product')
pivot2
```
