```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == '(':
            check = True
        if check == True:
            data = data + i
        if i == ')':
            break
    return data
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]
            return data
                
```

```python
df['Total_Time'] = df['Title'].apply(extraction_time)
```

```python
def extract_total_months(df, column_name, new_column='Total_Months'):
    # Split start and end date parts
    df[['Start', 'End']] = df[column_name].str.split(' - ', expand=True)

    # Convert text to datetime
    df['Start'] = pd.to_datetime(df['Start'], format='%b %Y', errors='coerce')
    df['End'] = pd.to_datetime(df['End'], format='%b %Y', errors='coerce')

    # Calculate total months difference
    df[new_column] = (
        (df['End'].dt.year - df['Start'].dt.year) * 12 +
        (df['End'].dt.month - df['Start'].dt.month)
    )

    # Optionally clean up intermediate columns
    df.drop(columns=['Start', 'End'], inplace=True)

    return df
```

```python
df = extract_total_months(df,'Total_Time')
df
```
