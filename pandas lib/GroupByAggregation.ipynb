```python
import numpy as np 
import pandas as pd
```

GROUPBY

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'], 
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,300,180,130],
    'Quantity':[10,15,12,10,18,8,25,6],
    'Date':pd.date_range('2025-10-29',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by category and Calculate sum of sales
```

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
stor = df.groupby('Store')['Sales'].sum()
stor
```

```python
stor = df.groupby(['Category','Store'])['Sales'].sum()
# stor
```
